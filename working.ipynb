```python
import os as os
import sys as sys
import re as re
import json as json
import urllib as urllib
import requests
import pandas as pd
```

```python
query_string = {
    'reviewed':'true',
    'compressed': 'false',
    'forma': 'json',
    'query' : f"""(gene:Rnf10) AND (taxonomy_id:10090) AND (reviewed:true)""",
    # 'fields': 'sequence'
}

query_string = urllib.parse.urlencode(query_string, quote_via=urllib.parse.quote)

url = f"https://rest.uniprot.org/uniprotkb/search?{query_string}"

response = requests.get(url)
```
